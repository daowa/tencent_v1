```python
import pandas as pd
import numpy as np
data = pd.read_csv("E:\\work\\kaggle\\data\\tecent\\pre\\position.csv")
data.head()
```

```python
dataTrain = pd.read_csv('E:\\work\\kaggle\\data\\tecent\\pre\\train.csv')
data = pd.merge(data, dataTrain, on='positionID')
```

```python
def click(field):
    return data[['label',field]].groupby(field).sum()
```

```python
def cvr(field):
    return data[['label',field]].groupby(field).sum()/data[['label',field]].groupby(field).count()
```

```python
click_data = click('positionID')
click_data
```

```python
cvr_data = cvr('positionID')
sort_cvr_data = cvr_data.sort_values(['label'], ascending = 0)['label']
sort_cvr_data
```

```python
import matplotlib.pyplot as plt
plt.plot(range(sort_cvr_data.size), sort_cvr_data)
plt.show()
```

```python
import math
sort_click_data = click_data.sort_values(['label'], ascending = 0)['label']
plt.plot(range(sort_click_data.size), sort_click_data.apply(lambda x: math.log(x+0.01)))
plt.show()
```
